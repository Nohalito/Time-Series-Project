```python
import os

IGNORE = {'.git', '__pycache__', '.RData', '.Rhistory'}

def print_tree(startpath, prefix=""):
    """
    Quick function to print our repositorie tree
    <3 ChatGPT for this one
    (`tree -I ".git|__pycache__"` doesn't work on my terminal, I guess window is at cause)
    """
    
    # List all items except ignored ones
    items = [i for i in os.listdir(startpath) if i not in IGNORE]
    
    # Separate files and directories
    files = sorted([i for i in items if os.path.isfile(os.path.join(startpath, i))])
    dirs = sorted([i for i in items if os.path.isdir(os.path.join(startpath, i))])
    
    # Combine files first, then directories
    sorted_items = files + dirs
    
    for idx, item in enumerate(sorted_items):
        path = os.path.join(startpath, item)
        connector = "└── " if idx == len(sorted_items) - 1 else "├── "
        print(prefix + connector + item)
        if os.path.isdir(path):
            extension = "    " if idx == len(sorted_items) - 1 else "│   "
            print_tree(path, prefix + extension)

# Usage: replace with your local repo path
print_tree("my/path/that/contain/personal/informations/so/I/hided/it")
```
